# Análise Limpa: Estado vs Crescimento

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

plt.rcParams["figure.figsize"] = (8,5)
```

## Carregar dados limpos

```
df = pd.read_csv("dataset_clean.csv")
df_brasil = pd.read_csv("dataset_brasil_clean.csv")

df.head()
```

## Estrutura

```
df.info()
```

## Estado vs Crescimento (por país)

```
plt.figure()

for country in df["country"].unique():
    subset = df[df["country"] == country]
    plt.scatter(subset["gov_size"], subset["gdp_growth"], label=country)

plt.xlabel("Tamanho do Estado (% PIB)")
plt.ylabel("Crescimento do PIB")
plt.title("Estado vs Crescimento (por país)")
plt.legend()

plt.show()
```

## Clusterização (KMeans)

```
X = df[["gov_size", "gdp_growth"]].dropna()

kmeans = KMeans(n_clusters=3, random_state=42)
df["cluster"] = kmeans.fit_predict(X)

plt.figure()

plt.scatter(df["gov_size"], df["gdp_growth"])
plt.xlabel("Estado (% PIB)")
plt.ylabel("Crescimento")
plt.title("Clusters Econômicos")

plt.show()
```

## Brasil: Crescimento vs Estado

```
plt.figure()

plt.plot(df_brasil["year"], df_brasil["gdp_growth"], label="PIB")
plt.plot(df_brasil["year"], df_brasil["state_growth"], label="Estado")

plt.xlabel("Ano")
plt.ylabel("Crescimento (%)")
plt.title("Brasil: PIB vs Crescimento do Estado")
plt.legend()

plt.show()
```
