# Análise Econômica: Estado vs Crescimento

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

## Carregar dados

```python
df_global = pd.read_csv("dataset_global.csv")
df_brasil = pd.read_csv("dataset_brasil.csv")

df_global.head()
```

## Estrutura dos dados

```python
df_global.info()
```

## Limpeza básica

```python
df_global = df_global.dropna(subset=["gdp_growth", "state_size"])
df_global = df_global.sort_values(["country", "year"])
```

## Estado vs Crescimento

```python
plt.figure()
sns.scatterplot(data=df_global, x="state_size", y="gdp_growth")
plt.title("Estado (% PIB) vs Crescimento")
plt.show()
```

## Correlação

```python
df_global[["state_size", "gdp_growth"]].corr()
```

## Regressão Linear

```python
import statsmodels.api as sm

X = df_global["state_size"]
y = df_global["gdp_growth"]

X = sm.add_constant(X)

model = sm.OLS(y, X).fit()
model.summary()
```

## Brasil

```python
plt.figure()
sns.lineplot(data=df_brasil, x="year", y="gdp_growth")
plt.title("Crescimento do PIB - Brasil")
plt.show()
```
