# Análise Econométrica: Estado vs Crescimento

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf

plt.rcParams["figure.figsize"] = (8,5)
```

## Carregar dados

```
df = pd.read_csv("dataset_econometria.csv")
df.head()
```

## Estrutura dos dados

```
df.info()
```

## Visual inicial

```
plt.figure()
plt.scatter(df["gov_lag1"], df["gdp_future"])
plt.xlabel("Estado (t-1)")
plt.ylabel("Crescimento futuro")
plt.title("Estado vs Crescimento Futuro")
plt.show()
```

## Regressão simples

```
model_simple = smf.ols(
    "gdp_future ~ gov_lag1",
    data=df
).fit()

print(model_simple.summary())
```

## Regressão com controles

```
model_controls = smf.ols(
    "gdp_future ~ gov_lag1 + investment + inflation + trade_open",
    data=df
).fit()

print(model_controls.summary())
```

## Modelo com efeitos fixos (país)

```
model_fe = smf.ols(
    "gdp_future ~ gov_lag1 + investment + inflation + trade_open + C(country)",
    data=df
).fit()

print(model_fe.summary())
```

## Linha de tendência

```
z = np.polyfit(df["gov_lag1"], df["gdp_future"], 1)
p = np.poly1d(z)

plt.figure()
plt.scatter(df["gov_lag1"], df["gdp_future"])
plt.plot(df["gov_lag1"], p(df["gov_lag1"]))
plt.title("Tendência")
plt.show()
```
